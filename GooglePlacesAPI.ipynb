```python
lat=float(input('Enter the latitude of your location in degrees: '))
long=float(input('Enter the longitude of your location in degrees: '))
r=3959.87

def degree_converter(a):
    return a*(3.14/180)

def distance_calculator(i,j,k,l,r):
    sines=math.sin(i)*math.sin(j)
    cosines=math.cos(i)*math.cos(j)*math.cos(k-l)
    arccosine=math.acos(sines+cosines)
    distance_sm=r*arccosine
    return distance_sm

def distance_converter(sm):
    return sm*1.609344

lat=degree_converter(lat)
long=degree_converter(long)

options=[]
for i in range(len(data)):
    sm=distance_calculator(lat,degree_converter(data[i][1]),long,degree_converter(data[i][2]),r)
    km=distance_converter(sm)
    options.append(km)

best=options.index(min(options))
address=addresses.index(data[best][0])

print('The closest hospital to you is {0}, located at {1}, {2:.2f}km away.'.format(data[best][0],addresses[address+1],min(options)))
```
